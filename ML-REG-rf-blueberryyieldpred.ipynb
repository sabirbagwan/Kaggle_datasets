```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler

from sklearn.ensemble import RandomForestRegressor
```

```python
data = pd.read_csv('/kaggle/input/wild-blueberry-yield-prediction/Data in Brief/Data in Brief/WildBlueberryPollinationSimulationData.csv')
data
```

```python
def preprocess_inputs(df):
    df = df.copy()
    
    ## Drop Row# column
    df = df.drop('Row#', axis = 1)
    
    ## Split the df into X and y
    y = df['yield']
    X = df.drop('yield', axis = 1)
    
    ## Train-Test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, shuffle = True, random_state = 1)
    
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test = preprocess_inputs(data)
X_train
```

```python
X_train.describe()
```

```python
model = RandomForestRegressor(random_state = 1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

rmse = np.sqrt(np.mean((y_test - y_pred)**2))
rmse
r2 = 1 - (np.sum((y_test - y_pred)**2) / np.sum((y_test - y_test.mean())**2))
r2
# y_pred
print(f'RMSE: {rmse} \nR^2: {r2}')
```

```python
params = {
   'n_estimators' : [50, 100, 150, 200],
    'max_depth' : [2, 4, 6, 8, 10]
}

model = GridSearchCV(RandomForestRegressor(random_state = 1), params)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

rmse = np.sqrt(np.mean((y_test - y_pred)**2))
rmse
r2 = 1 - (np.sum((y_test - y_pred)**2) / np.sum((y_test - y_test.mean())**2))
r2
print(f'RMSE: {rmse} \nR^2: {r2}')
```
